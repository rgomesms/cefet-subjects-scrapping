```python
import requests
from bs4 import BeautifulSoup
```

```python
html_page = requests.get('http://www.decom.cefetmg.br/ensino/graduacao/engenharia-de-computacao/disciplinas/').text
```

```python
bs = BeautifulSoup(html_page, 'html.parser')
```

```python
for link in bs.find_all('a'):
    cur_href = link.get('href')
    if isinstance(cur_href, str) and cur_href.endswith('.pdf'):
        print(link.get('title'))
```
